```python
import numpy as np
from Plotting.Plot import *
```

```python
%matplotlib inline
phi = np.linspace(-np.pi, 3*np.pi, 200)
r_2 = 0.9
a = 1
r_1 = 0.9

def TDrop(phi):
    return (((r_2*a)**2 - 2*r_1*r_2*np.cos(phi) + r_1**2) / 
           (1 - 2*r_1*r_2*a*np.cos(phi) + (r_1*r_2*a)**2))

def TPass(phi):
    return (((1-r_1**2)*(1-r_2**2)*a) / 
           (1 - 2*r_1*r_2*a*np.cos(phi)+(r_1*r_2*a)**2))

T_d = TDrop(phi)
T_p = TPass(phi)

plot = MultiPlot(
    [phi / np.pi ,phi / np.pi],
    [T_d,T_p],
    'Detuning',
    'Transmission',
    ['Drop', 'Pass'])

plot.setup_plot()
plot.ax.xaxis.set_major_formatter(plt.FormatStrFormatter('%g $\pi$'))
plot.ax.xaxis.set_major_locator(mpl.ticker.MultipleLocator(base=1.0))
plt.savefig('TPass_TDrop' + ".pdf", format="pdf", dpi=1200, bbox_inches="tight")
```

```python
x = np.linspace(-1,1, 100)
y = np.maximum(0,x)

plot = BasicPlot(
    x,
    y,
    'x',
    'y')

plot.setup_plot()
plot.save('ReLU_Function')
```

```python
alpha = 0.1
g_phi = 2.2
phi_b = np.array([0, np.pi/6, np.pi/4, np.pi/3, np.pi/2, 
                 3*np.pi/4, 2*np.pi/3, 5*np.pi/6, np.pi])[:,np.newaxis]
norm = np.linalg.norm

def f(z):
    arg = (g_phi*np.abs(z)**2 / 2 + phi_b/2)
    return np.abs(1j*np.sqrt(1-alpha)*(np.cos(arg) - 1j*np.sin(arg))*np.cos(arg)*z)

def phi(z):
    arg = (g_phi*np.abs(z)**2 / 2 + phi_b/2)
    return np.angle(1j*np.sqrt(1-alpha)*(np.cos(arg) - 1j*np.sin(arg))*np.cos(arg)*z)

scale = np.sqrt(g_phi/np.pi)
z = np.linspace(1e-8,1, 50)[np.newaxis,:]
t = (f(z*scale))**2 / z**2
t = t / np.max(t)

phase = phi(z*scale)

for y,lab in zip(t,phi_b):
    plt.plot(z[0],y, label=lab[0])
plt.legend()
plt.show()
for y,lab in zip(phase,phi_b):
    plt.plot(z[0],y, label=lab[0])
plt.legend()
```
